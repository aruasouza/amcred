```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_con = pd.read_feather('BF_Dados_Contratos.feather')
```

```python
df_socio = pd.read_excel('BF_Dados_Socio_Economicos.xlsx')
```

```python
df_con.info()
```

```python
df_con.describe()
```

```python
numeric = ['RENDA_CLIENTE','QUANTIDADE_PARCELAS', 'QUANTIDADE_PAGAS','VALOR_EMPRESTIMO',
       'PRESTACAO', 'ATRASO_DIAS', 'ATRASO_MAXIMO', 'RENEGOCIADO']
corr = df_con[numeric].corr()
corr
```

```python
for i in range(len(corr)):
    corr.iloc[i,i] = 0
    corr.iloc[i] = corr.iloc[i].apply(abs)
corr
```

```python
f = plt.figure(figsize=(15, 10))
plt.matshow(df_con[numeric].corr(),fignum=f.number)
plt.xticks(range(len(numeric)), numeric, fontsize=14, rotation=45)
plt.yticks(range(len(numeric)), numeric, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)
```
